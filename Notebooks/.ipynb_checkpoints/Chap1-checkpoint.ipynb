```python
import matplotlib.pyplot as plt
# get from the my_plots python file custom parametrization for matplotlib
# get_custom_colors is a set of colorblind-friendly colors (pour les daltoniens!!)
from my_plots import set_size, get_custom_colors

import matplotlib as mpl
from cycler import cycler
```

```python
# import useful libraries for my work
import numpy as np
import pandas as pd
```

```python
fig_dir = '../Manuscript/fichiers_latex/Chap1/Figure/'
```

```python
# this cycler will be implemented in matplotlib customized parameters
# it makes colors of plots change automatically in the defined order
custom_cycler = cycler(color=get_custom_colors())
```

```python
# textwidth of the latex thesis manuscript (in points)
textwidth = 426.79134
```

```python
custom_fonts = {
        # Use LaTeX to write all text
        "text.usetex": True,
        "font.family": "serif",
        # Use 10pt font in plots, to match 10pt font in document
        "axes.labelsize": 10,
        "font.size": 10,
        # Make the legend/label fonts a little smaller
        "legend.fontsize": 8.5,
        "xtick.labelsize": 8.5,
        "ytick.labelsize": 8.5,
        # color cycler
        "axes.prop_cycle": custom_cycler,
        # auto tight layout
        'figure.autolayout' : True,
        # hatch line widths
        'hatch.linewidth':2.,
        #titles too in 10 nt
        'axes.titlesize' : 10,
        # savefig with tight bbox_inches
        'savefig.bbox' : 'tight'
}

mpl.rcParams.update(custom_fonts)
```

```python
# line and color styles for "by hand" changes
linestyles = 2*['-', '--', '--', '-.', ':','-']
colors = 2 * get_custom_colors()
```

# Les figures

By changing the settings of the figures here directly, each figure is saved in the appropriate directory (fig_dir), which may be called directly by latex... aka directly updated in your manuscript ;-)

## Example of a plot, directly updated in manuscript

```python
x = np.linspace(- 2 * np.pi, 2 * np.pi,100)
```

```python
fig,ax = plt.subplots(1,1,
                     figsize=set_size(textwidth))

for fs in range(10):
    ax.plot(x,np.sin(x)+fs,
            lw=2)
plt.savefig(fig_dir + 'Figure_simple_largeur_max.pdf')
```

## With different linestyles and smaller size option

```python
fig,ax = plt.subplots(1,1,
                     figsize=set_size(textwidth,fraction=0.7))

for fs in range(10):
    ax.plot(x,np.sin(x)+fs,
            lw=2,
           ls=linestyles[fs])

ax.set_xlabel('Font size is great.')

ax.set_title('Title size looks good !')

plt.savefig(fig_dir + 'Figure_simple_fraction.pdf')
```

## Higher but same width option

```python
fig,ax = plt.subplots(1,1,
                     figsize=set_size(textwidth,shrink_height=1.5))

for fs in range(4):
    ax.plot(x,np.sin(x)+fs,
            lw=2)
plt.savefig(fig_dir + 'Figure_simple_larger_height.pdf')
```

## Or smaller in height option

```python
fig,ax = plt.subplots(1,1,
                     figsize=set_size(textwidth, shrink_height=0.5))

for fs in range(4):
    ax.plot(x,np.sin(x)+fs,
            lw=2)
plt.savefig(fig_dir + 'Figure_simple_smaller_height.pdf')
```

## Finally, mulitple subplotswith matplotlib

```python
fig,ax = plt.subplots(2,1,
                     figsize=set_size(textwidth, subplot=[2,1]))

for fs in range(4):
    ax[0].plot(x,np.sin(x)+fs,
            lw=2)
    ax[1].plot(x,np.cos(2*x)+fs,
            lw=2,
              label=f'fs : {fs:.0f}')

ax[1].legend(loc='upper right')
ax[1].set_xlabel('Font size of legend looks fine too.')

plt.savefig(fig_dir + 'Figure_multiple_21.pdf')
```

```python
fig,ax = plt.subplots(1,2,
                     figsize=set_size(textwidth, subplot=[1,2]))

for fs in range(4):
    ax[0].plot(x,np.sin(x)+fs,
            lw=2)
    ax[1].plot(x,np.cos(2*x)+fs,
            lw=2)

plt.savefig(fig_dir + 'Figure_multiple_12.pdf')
```
